```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the provided Excel file
df = pd.read_excel(r'C:\Users\Lenovo\Downloads\diurinal.xlsx')

# Convert 'From Date' to datetime format
df['From Date'] = pd.to_datetime(df['From Date'], format='%d-%m-%Y %H:%M')

# Extract hour from 'From Date'
df['Hour'] = df['From Date'].dt.hour

# Group by the hour and calculate mean for each pollutant
hourly_avg = df.groupby('Hour').mean()

# Set up the plot size
plt.figure(figsize=(12, 8))

# List of pollutants to plot
pollutants = ['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'SO2', 'CO', 'Ozone', 'Benzene']

# Plot each pollutant
for pollutant in pollutants:
    plt.plot(hourly_avg.index, hourly_avg[pollutant], label=pollutant)

# Adding labels and title
plt.title('Hourly Average Concentrations of Pollutants (Diurnal Graph)', fontsize=16)
plt.xlabel('Hour of the Day', fontsize=12)
plt.ylabel('Concentration (µg/m³)', fontsize=12)
plt.xticks(range(0, 24, 1))  # Showing every hour
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the data from Excel
# Replace 'your_file.xlsx' with your actual file path
df = pd.read_excel('your_file.xlsx')

# Step 2: Ensure that the timestamp is in datetime format if it's not already
df['Timestamp'] = pd.to_datetime(df['Timestamp'])

# Step 3: Extract the hour from the timestamp
df['Hour'] = df['Timestamp'].dt.hour

# Step 4: Group by the hour and calculate the average pollutant level
# Replace 'Pollutant_Level' with the actual name of the column containing pollutant data
hourly_avg = df.groupby('Hour')['Pollutant_Level'].mean()

# Step 5: Plot the diurnal variation graph
plt.plot(hourly_avg.index, hourly_avg.values, marker='o')
plt.title('Diurnal Variation of Pollutant Levels')
plt.xlabel('Hour of the Day')
plt.ylabel('Average Pollutant Level')
plt.xticks(range(0, 24))  # X-axis labels for each hour of the day
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the Excel file
file_path = r'C:\Users\Lenovo\Downloads\diurinal.xlsx'
df = pd.read_excel(file_path)

# Step 2: Convert 'From Date' to datetime format
df['From Date'] = pd.to_datetime(df['From Date'], format='%d-%m-%Y %H:%M')

# Step 3: Extract the hour from 'From Date'
df['Hour'] = df['From Date'].dt.hour

# Step 4: List of pollutant columns
pollutants = ['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'SO2', 'CO', 'Ozone', 'Benzene']

# Step 5: Plot the diurnal variation for each pollutant
for pollutant in pollutants:
    hourly_avg = df.groupby('Hour')[pollutant].mean()
    plt.plot(hourly_avg.index, hourly_avg.values, marker='o', label=pollutant)

# Step 6: Customize the plot
plt.title('Diurnal Variation of Pollutants')
plt.xlabel('Hour of the Day')
plt.ylabel('Average Pollutant Level')
plt.xticks(range(0, 24))  # X-axis labels for each hour of the day
plt.legend(title='Pollutants', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```
